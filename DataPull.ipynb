```python
# use glob to get all the csv files 
# in the folder
path = os.getcwd()
csv_files = glob.glob(os.path.join("./Resources", "JC-*.csv"))
  
all_data = []  
#loop over the list of csv files
for f in csv_files:
      
    # read the csv file
    df = pd.read_csv(f)
   
      
    # print the location and filename
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])
      
    # print the content
    print('Content:')
    display(df)
all_data.append(df)
print(all_data)
```
